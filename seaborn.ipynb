```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

Seaborn Bar Plot

1. Write a program to draw bar plot of sex against survived for a dataset given in the url

```python
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv")
sns.countplot(data=df, x="sex", hue="survived")

plt.xlabel("Sex")
plt.ylabel("Count")
plt.title("Survival Count by Gender")

plt.show()
```

Seaborn Pointplot

1. Write a program to draw a point plot for sex against survived for a dataset given in url

```python
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv")
sns.pointplot(data=df, x="sex", y="survived", capsize=0.2, markers="o", color="blue")

plt.xlabel("Sex")
plt.ylabel("Survival Rate")
plt.title("Survival Rate by Gender")

plt.show()
```

Seaborn scatterplot

1. Write a program to draw a scatter plot of “day” against “total bill” for a dataset given in a
url


```python
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")
sns.scatterplot(data=df, x="day", y="total_bill", alpha=1)

plt.xlabel("Day of the Week")
plt.ylabel("Total Bill ($)")
plt.title("Total Bill Distribution by Day")

plt.show()
```

Seaborn Violin Plot

1. Write a program to draw a violin plot of sex against total_bill for a given dataset

```python
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")

plt.figure(figsize=(8, 6)) 
sns.violinplot(data=df, x="sex", y="total_bill", inner="quartile", hue="sex", palette="coolwarm")
plt.title("Distribution of Total Bill by Gender")
plt.show()
```

2. Write a program to draw a violin plot of “species” against “sepal length” for a dataset
given in a url


```python
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv")

sns.violinplot(data=df, x="species", y="sepal_length", hue = "species",inner="quart", palette="Set2")

plt.xlabel("Species")
plt.ylabel("Sepal Length (cm)")
plt.title("Distribution of Sepal Length by Species")

plt.show()
```

Seaborn BoxPlot

1. Write a program to draw box plot of life expectancy by continent for a data set given in a
url


```python
df = pd.read_csv("https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapminder-FiveYearData.csv")
sns.boxplot(data=df, x="continent", y="lifeExp",hue = 'year', palette="Set2")

plt.xlabel("Continent")
plt.ylabel("Life Expectancy")
plt.title("Life Expectancy by Continent")
plt.show()
```
